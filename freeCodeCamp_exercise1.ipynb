### Add and calculate a new `Calculated_Date` column

Use `Day`, `Month`, `Year` to create a `Date` column (`YYYY-MM-DD`).

```python
# first attempt
# sales['Year'].astype(str)+"-"+pd.to_datetime(sales['Month'], format="%B").dt.month.astype(str).str.zfill(2)+"-"+sales['Day'].astype(str).str.zfill(2)
```

```python
# better
sales['Calculated_Date'] = pd.to_datetime({
    'year' : sales['Year'],
    'month' : pd.to_datetime(sales['Month'], format="%B").dt.month,
    'day' : sales['Day']
}).dt.strftime("%Y-%m-%d")
# .dt onward is optional

sales['Calculated_Date'].head()
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)
